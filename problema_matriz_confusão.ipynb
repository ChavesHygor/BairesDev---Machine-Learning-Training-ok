```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
# --- ETAPA 1: DADOS DE EXEMPLO ---
# Usaremos os mesmos dados do exemplo anterior para consistência.
# Classe 1: Positivo | Classe 0: Negativo

# y_true: Os valores verdadeiros (gabarito)
y_true = [1, 0, 1, 1, 0, 1, 0, 0, 1, 0, 1, 1]

# y_pred: Os valores que o nosso modelo previu
y_pred = [1, 1, 1, 0, 0, 1, 0, 0, 1, 0, 0, 1]
```

```python
# --- ETAPA 2: GERAR E PLOTAR A MATRIZ DE CONFUSÃO ---

# Primeiro, geramos a matriz numericamente com o scikit-learn
cm = confusion_matrix(y_true, y_pred)

print("--- Matriz de Confusão Visualizada ---")

# Agora, vamos plotar essa matriz usando Seaborn e Matplotlib
plt.figure(figsize=(8, 6)) # Define o tamanho da figura

# Cria o heatmap
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Previsto Negativo (0)', 'Previsto Positivo (1)'],
            yticklabels=['Real Negativo (0)', 'Real Positivo (1)'])

# Adiciona títulos e rótulos para maior clareza
plt.title('Matriz de Confusão', fontsize=16)
plt.ylabel('Valores Reais (Gabarito)', fontsize=12)
plt.xlabel('Valores Previstos pelo Modelo', fontsize=12)
plt.show() # Exibe o gráfico
```

```python
# --- ETAPA 3: LEITURA DA MATRIZ E CÁLCULO DAS MÉTRICAS ---
print("\n--- Leitura dos Dados da Matriz Plotada ---")

# Agora, vamos extrair os valores dessa mesma matriz que acabamos de visualizar
# para fazer os cálculos. A estrutura é a mesma:
# [[VN, FP],
#  [FN, VP]]

VN = cm[0, 0]
FP = cm[0, 1]
FN = cm[1, 0]
VP = cm[1, 1]

print(f"Verdadeiros Positivos (VP): {VP}")
print(f"Verdadeiros Negativos (VN): {VN}")
print(f"Falsos Positivos (FP): {FP}  <-- Erro Tipo I")
print(f"Falsos Negativos (FN): {FN}   <-- Erro Tipo II")
print("-" * 40)

# N: Total de elementos
N = VP + VN + FP + FN
```

```python
# Calculando as métricas com base nos valores lidos
sensibilidade = VP / (VP + FN)
especificidade = VN / (FP + VN)
acuracia = (VP + VN) / N
precisao = VP / (VP + FP)
f_score = 2 * (precisao * sensibilidade) / (precisao + sensibilidade)


# --- ETAPA 4: EXIBIR OS RESULTADOS FINAIS ---

print("\n--- Métricas de Avaliação Calculadas ---")
print(f"Sensibilidade (Recall): {sensibilidade:.4f}  ({sensibilidade:.2%})")
print(f"Especificidade:         {especificidade:.4f}  ({especificidade:.2%})")
print(f"Acurácia:               {acuracia:.4f}  ({acuracia:.2%})")
print(f"Precisão:               {precisao:.4f}  ({precisao:.2%})")
print(f"F-score:                {f_score:.4f}  ({f_score:.2%})")
print("-" * 40)
print(f"Total de elementos (N): {N}")
```
